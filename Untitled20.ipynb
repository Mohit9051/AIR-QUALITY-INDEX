```python
mod_air_quality_data_in_india = air_quality_data_in_india.dropna(subset = ['AQI'])
mod_air_quality_data_in_india
```

```python
mod_air_quality_data_in_india = mod_air_quality_data_in_india.dropna(thresh = 8, subset = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene'])
mod_air_quality_data_in_india
```

```python
pol_col = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene']
for col in pol_col:
    mod_air_quality_data_in_india[col] = mod_air_quality_data_in_india.groupby(['City'])[col].transform(lambda x: np.where(np.isnan(x), x.mean(),x))
```

```python
mod_air_quality_data_in_india[['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI']] = mod_air_quality_data_in_india[['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI']].fillna(value = 0)
mod_air_quality_data_in_india
```

```python
from sklearn.preprocessing import minmax_scale
from sklearn.preprocessing import OrdinalEncoder
```

```python
#print("Before outlier")
#print(mod_air_quality_data_in_india['NH3'].tail())
def is_outlier(s):
    lower_limit = s.mean() - (s.std() * 2)
    upper_limit = s.mean() + (s.std() * 2)
    return ~s.between(lower_limit, upper_limit)

pol_col = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI']
for col in pol_col:
    feat_enggd = mod_air_quality_data_in_india[~mod_air_quality_data_in_india.groupby('City')[col].apply(is_outlier)]

#print("After outlier")
#print(mod_air_quality_data_in_india['NH3'].tail())
```

```python
sns.set(rc = {'figure.figsize':(25,25)})
ax = sns.boxplot(x = 'AQI', y = 'City', data = feat_enggd)
```

```python
pol_col = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI']
for col in pol_col:
    feat_enggd = mod_air_quality_data_in_india.copy()
    feat_enggd[col] = feat_enggd.groupby(['City'])[col].transform(lambda x: minmax_scale(x.astype(float)))
feat_enggd
```

```python
from datetime import datetime as dt
from datetime import date
enc = OrdinalEncoder()
feat_enggd[["AQI_Bucket"]] = enc.fit_transform(feat_enggd[["AQI_Bucket"]])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
city_col = ['Bengaluru']
i=0
city_AQ = {}
for col in city_col:
    df = feat_enggd.groupby('City')
    city_AQ[i] = df.get_group(col)
    i=i+1
```

```python
city_col = ['Bengaluru']
i=0
for col in city_col:
    DataFrame = city_AQ[i].copy()
    O_Date = pd.to_datetime(DataFrame['Date']).apply(lambda date: date.toordinal())
    X = O_Date
    y = DataFrame['AQI']
    X = X.values.reshape(-1, 1)    
    y = y.values.reshape(-1, 1)
        
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
    lr_model = LinearRegression()
    lr_model.fit(X_train, y_train)
    y_predict = lr_model.predict(X_test)
    mse = mean_squared_error(y_test, y_predict)
    r2 = r2_score(y_test, y_predict)
    mae = mean_absolute_error(y_test, y_predict)
    print('----------------------------------------')
    print(f'{col}')
    print('mean_squared_error : ',mse)
    print('r2_score           : ',r2)
    print('mean_absolute_error: ',mae)
    print('----------------------------------------')
    ticks = ['2015','2016','2017','2018','2019','2020']
    xtickpos = np.linspace(min(O_Date), max(O_Date), 6)
    ax = plt.subplot(7,4,i+1)
    plt.scatter(X, y, c = 'yellow', alpha = 0.12)
    plt.scatter(X_test, y_test, c = 'green', alpha = 0.24)
    plt.plot(X_test, y_predict, c = 'red')
    plt.xticks(xtickpos, ticks)
    plt.tight_layout()
    ax.set_title(f'{col}')
    i=i+1
```
